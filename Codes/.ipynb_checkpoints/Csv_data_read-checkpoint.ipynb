```python
from numba import jit


@jit
def run_vid(cap):
    pixVal = []
    pixVal2 = []
    cropmn = []
    loop_t = 0
    while(cap.isOpened()):
        ret, frame = cap.read() 

        if ret ==0:
            break
        # converting BGR to HSV 

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(
                gray,
                scaleFactor=1.1,
                minNeighbors=5,
                minSize = (30,30),
                flags = cv2.CASCADE_SCALE_IMAGE
                )

        cropIm = frame[600:700, 700:750]
        cropmn.append(cropIm.mean())

        #eyes = eye_cascade.detectMultiScale(gray, 1.1,3)

        for f in faces:
            x, y, w, h = [ v for v in f ]
           # cv2.rectangle(frame, (x,y), (x+w,y+h), (255,255,255))

            sub_face = frame[y-50:y+h+50, x:x+w]
            
            sub_face2 = cv2.resize(sub_face, (600,700))
            
            cv2.rectangle(sub_face2, (50,300), (250, 500), (255,255,255))

        # Display the resulting frame
        
        cv2.imshow('Video', sub_face2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        loop_t = +1
        
    pixd = {'p1': pixVal, 'p2': pixVal2, 'cpm' : cropmn}
            
    return pixd
"""
def main():
    cap = cv2.VideoCapture('p13_normal.MP4')
    # Check if camera opened successfully
    pixVal = run_vid(cap)
    cap.release()
    cv2.destroyAllWindows()


if __name__ == "__main__":
    main()
    """
```
